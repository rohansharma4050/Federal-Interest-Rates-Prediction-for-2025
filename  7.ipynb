```python
#!/usr/bin/env python
# coding: utf-8

""""

Objective:
Train a Convolutional Neural Network (CNN) to classify images and identify an intentional dimensionality error in the model architecture. 
Correct the error to ensure the model runs successfully and produces accurate predictions.

Instructions:
1. Implement the 'compute_conv2d_output_dim' and 'compute_pool2d_output_dim' functions.
2. Run the script to download the dataset, preprocess the data, build and train the CNN model.
3. Identify and correct the dimensionality error in the MaxPooling layer.
4. Ensure the corrected code runs without errors and achieves accurate predictions.
"""

import os
import sys
import zipfile
import subprocess
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

# Suppress TensorFlow warnings for cleaner output
import logging
logging.getLogger('tensorflow').setLevel(logging.ERROR)

# ==========================
# 1. Dimension Calculation Functions
# ==========================

def compute_conv2d_output_dim(input_size, filter_size, stride=1, padding='valid'):
    """
    Computes the output dimensions of a Conv2D layer.

    Parameters:
        input_size (int): Size of the input (height or width).
        filter_size (int): Size of the filter (height or width).
        stride (int): Stride of the convolution.
        padding (str): 'valid' or 'same'.

    Returns:
        int: Output size after convolution.
    """
    if padding == 'same':
        output_size = int(np.ceil(float(input_size) / float(stride)))
    elif padding == 'valid':
        output_size = int(np.floor((input_size - filter_size + stride) / stride))
    else:
        raise ValueError("Padding must be 'same' or 'valid'")
    return output_size

def compute_pool2d_output_dim(input_size, pool_size=2, stride=2, padding='valid'):
    """
    Computes the output dimensions of a Pooling2D layer.

    Parameters:
        input_size (int): Size of the input (height or width).
        pool_size (int): Size of the pooling window.
        stride (int): Stride of the pooling operation.
        padding (str): 'valid' or 'same'.

    Returns:
        int: Output size after pooling.
    """
    if padding == 'same':
        output_size = int(np.ceil(float(input_size) / float(stride)))
    elif padding == 'valid':
        output_size = int(np.floor((input_size - pool_size + stride) / stride))
    else:
        raise ValueError("Padding must be 'same' or 'valid'")
    return output_size

# ==========================
# 2. Download and Extract Dataset
# ==========================

def download_and_extract_dataset():
    """
    Downloads and extracts the CIFAR-10 dataset using TensorFlow Keras API.

    Returns:
        tuple: (train_data, test_data)
    """
    print("Downloading and loading CIFAR-10 dataset...")
    (train_data, train_labels), (test_data, test_labels) = tf.keras.datasets.cifar10.load_data()
    print("Dataset loaded successfully.")
    return (train_data, train_labels), (test_data, test_labels)


# ==========================
# 3. Prepare Data
# ==========================

def prepare_data(train_data, train_labels, test_data, test_labels):
    """
    Prepares ImageDataGenerators for training and validation.

    Parameters:
        train_data (numpy.ndarray): Training images.
        train_labels (numpy.ndarray): Training labels.
        test_data (numpy.ndarray): Testing images.
        test_labels (numpy.ndarray): Testing labels.

    Returns:
        tuple: (train_generator, validation_generator, num_classes)
    """
    # Normalize pixel values
    train_data = train_data.astype('float32') / 255.0
    test_data = test_data.astype('float32') / 255.0

    # Convert labels to one-hot encoding
    num_classes = 10
    train_labels = tf.keras.utils.to_categorical(train_labels, num_classes)
    test_labels = tf.keras.utils.to_categorical(test_labels, num_classes)
    
    # Image dimensions
    IMG_HEIGHT, IMG_WIDTH = train_data.shape[1], train_data.shape[2]

    # Create ImageDataGenerators
    train_datagen = ImageDataGenerator(
        rotation_range=15,
        width_shift_range=0.1,
        height_shift_range=0.1,
        horizontal_flip=True
    )

    test_datagen = ImageDataGenerator()

    # Create generators
    train_generator = train_datagen.flow(
        train_data, train_labels,
        batch_size=64
    )

    validation_generator = test_datagen.flow(
        test_data, test_labels,
        batch_size=64
    )


    return train_generator, validation_generator, num_classes, IMG_HEIGHT, IMG_WIDTH, len(train_data), len(test_data)

# ==========================
# 4. Build the Neural Network
# ==========================

def build_cnn_model(input_shape, num_classes):
    """
    Builds and returns a simple CNN model.

    Parameters:
        input_shape (tuple): Shape of the input images (height, width, channels).
        num_classes (int): Number of output classes.

    Returns:
        tf.keras.Model: Compiled CNN model.
    """
    model = models.Sequential()

    # First Convolutional Layer
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape, name='Conv1'))
    model.add(layers.MaxPooling2D((2, 2), name='Pool1'))

    # Second Convolutional Layer
    model.add(layers.Conv2D(64, (3, 3), activation='relu', name='Conv2'))
    # Intentional Dimensionality Error: Using (3,3) instead of (2,2) for pooling
    model.add(layers.MaxPooling2D((2, 2), name='Pool2'))  # Error here # Corrected pooling size: My comment

    # Flatten and Dense Layers
    model.add(layers.Flatten(name='Flatten'))
    model.add(layers.Dense(64, activation='relu', name='Dense1'))
    model.add(layers.Dense(num_classes, activation='softmax', name='Output'))

    # Compile the Model
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    print("\nModel Summary:")
    model.summary()
    return model


# ==========================
# 5. Train the Model
# ==========================

def train_model(model, train_generator, validation_generator, epochs=10, train_samples=0, val_samples=0):
    """
    Trains the CNN model.

    Parameters:
        model (tf.keras.Model): The CNN model to train.
        train_generator (DirectoryIterator): Training data generator.
        validation_generator (DirectoryIterator): Validation data generator.
        epochs (int): Number of training epochs.

    Returns:
        History: Training history.
    """
    # Calculate the number of steps per epoch
    steps_per_epoch = train_samples // train_generator.batch_size
    validation_steps = val_samples // validation_generator.batch_size
    
    history = model.fit(
        train_generator,
        #steps_per_epoch=len(train_generator),  # Total batches in training data
        steps_per_epoch=steps_per_epoch,
        epochs=epochs,
        validation_data=validation_generator,
        #validation_steps=len(validation_generator)  # Total batches in validation data
        validation_steps=validation_steps
    )
    return history

# ==========================
# 6. Plot Training History
# ==========================

def plot_training_history(history):
    """
    Plots training and validation accuracy and loss.

    Parameters:
        history (History): Training history.
    """
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']

    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs_range = range(len(acc))

    plt.figure(figsize=(14, 6))

    # Plot Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy', marker='o')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy', marker='x')
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')
    plt.grid(True)

    # Plot Loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss', marker='o')
    plt.plot(epochs_range, val_loss, label='Validation Loss', marker='x')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.grid(True)

    plt.show()
    
    
# ==========================
# 7. Main Execution
# ==========================
def main():
    """
    Main function to execute the assignment steps.
    """
    # Step 1: Download and Extract Dataset
    try:
        (train_data, train_labels), (test_data, test_labels) = download_and_extract_dataset()
    except Exception as e:
        print(f"Error downloading or extracting dataset: {e}")
        sys.exit(1)

    # Step 2: Prepare Data
    try:
        train_generator, validation_generator, num_classes, IMG_HEIGHT, IMG_WIDTH, train_samples, val_samples = prepare_data(
            train_data, train_labels, test_data, test_labels)
    except Exception as e:
        print(f"Error preparing data: {e}")
        sys.exit(1)

    # Step 3: Build Model
    input_shape = (IMG_HEIGHT, IMG_WIDTH, 3)  # CIFAR-10 images are RGB
    model = build_cnn_model(input_shape, num_classes)

    # Step 4: Train Model
    EPOCHS = 10
    try:
        history = train_model(model, train_generator, validation_generator, epochs=EPOCHS, 
                              train_samples=train_samples, val_samples=val_samples)
    except Exception as e:
        print(f"Error during training: {e}")
        sys.exit(1)

    # Step 5: Plot Training History
    try:
        plot_training_history(history)
    except Exception as e:
        print(f"Error plotting training history: {e}")

if __name__ == "__main__":
    main()
```
